```python
from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('https://www.simplilearn.com/skillup-free-online-courses').text

soup= BeautifulSoup(source, 'html.parser')

with open('simplilearn_scrape.csv', 'w', encoding='utf-8', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['course name', 'duration', 'rating', 'advantage', 'course link'])

    for course in soup.find_all('div', class_='card-container'):
            coursename = course.find('div', class_='info-exp').text
            print(coursename)

            #rating = course.find('span', class_='rate').text
            #print(rating)

            duration = course.find('span', class_='duration').text
            print(duration)

            advantage= course.find('div', class_='completion-certificate').text
            print(advantage)

            link_course = course.find('a', class_='view')['href']

            view_course = link_course.split('/')[1]
            view_course = view_course.split('?')[0]

            link_view = f'https://www.simplilearn.com/{view_course}'
            print(link_view)


            try:
                rating = course.find('span', class_='rate').text

            except Exception as e:
                rating = None

            print(rating)

            print( )

            csv_writer.writerow([coursename, duration, rating,  advantage, link_view])

    #csv_file.close()
```
